```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# check sizes of results
df = pd.read_csv("cnn_results.csv")

print(df.groupby(["size_increase"]).count())
```

```python
plot_df = df.groupby(["original_size","size_increase","new_size"]).agg({"accuracy":"mean"}).reset_index()


for size_inc in plot_df['size_increase'].unique():
    i_df = plot_df[plot_df["size_increase"]==size_inc]
    plt.plot(i_df["original_size"],i_df["accuracy"],label=f"{1+size_inc}")

plt.legend()
plt.xlabel("Development Dataset size")
plt.ylabel("Accuracy")
plt.show()
```

```python
dense_df = pd.read_csv("dense_results.csv")
plot_df = dense_df.groupby(["original_size","size_increase","new_size"]).agg({"accuracy":"mean"}).reset_index()


for size_inc in plot_df['size_increase'].unique():
    i_df = plot_df[plot_df["size_increase"]==size_inc]
    plt.plot(i_df["original_size"],i_df["accuracy"],label=f"{1+size_inc}")

plt.legend()
plt.title("Dense Layer Classifier")
plt.xlabel("Development Dataset size")
plt.ylabel("Accuracy")
plt.show()
```

```python
scnn_df = pd.read_csv("simple_cnn_results.csv")
plot_df = scnn_df.groupby(["original_size","size_increase","new_size"]).agg({"accuracy":"mean"}).reset_index()


for size_inc in plot_df['size_increase'].unique():
    i_df = plot_df[plot_df["size_increase"]==size_inc]
    plt.plot(i_df["original_size"],i_df["accuracy"],label=f"{1+size_inc}")

plt.legend()
plt.title("Simple CNN Classifier")
plt.xlabel("Development Dataset size")
plt.ylabel("Accuracy")
plt.show()
```

```python
cnn_df = pd.read_csv("cnn_results.csv")
plot_df = cnn_df.groupby(["original_size","size_increase","new_size"]).agg({"accuracy":"mean"}).reset_index()


for size_inc in plot_df['size_increase'].unique():
    i_df = plot_df[plot_df["size_increase"]==size_inc]
    plt.plot(i_df["original_size"],i_df["accuracy"],label=f"{1+size_inc}")

plt.legend(loc=4)
plt.title("CNN Classifier")
plt.xlabel("Development Dataset size")
plt.ylabel("Accuracy")
plt.show()
```

```python
fdf = pd.read_csv("10000_fixed.csv")

models = list(fdf["model"].unique())
for model in models:
    i_df = fdf[fdf["model"]==model].groupby("original_size").agg({"accuracy":"mean"}).reset_index()
    plt.plot(i_df["original_size"],i_df["accuracy"],label=model)


plt.title("Fixed Development size: 10,000")
plt.xlabel("Real images count")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
