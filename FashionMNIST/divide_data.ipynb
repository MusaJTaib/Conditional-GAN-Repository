```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

physical_devices = tf.config.experimental.list_physical_devices('GPU')
print(physical_devices)
```

```python
(X_dev,y_dev),(X_test,y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
# shuffle data
indexes = np.arange(X_dev.shape[0],dtype=int)
np.random.shuffle(indexes)
X_dev = X_dev[indexes]
y_dev = y_dev[indexes]
```

```python
###### split by percentages
sample_counts = [1000,2000,3000,4000,5000,6000,7000,8000,9000,10000]

for sample_count in sample_counts:
    X_iter = X_dev[:sample_count]
    y_iter = y_dev[:sample_count]

    suffix = "_{}_data".format(sample_count)
    x_name = f"data/base_data/MNIST_X_dev{suffix}.npy"
    y_name = f"data/base_data/MNIST_y_dev{suffix}.npy"

    np.save(x_name,X_iter)
    np.save(y_name,y_iter)
```

```python
#save test dataset
x_test_name = "data/MNIST_X_test.npy"
y_test_name = "data/MNIST_y_test.npy"
np.save(x_test_name,X_test)
np.save(y_test_name,y_test)
```
